```python
#load the data
import pandas as pd 

#Load your CSV
df = pd.read_csv("sales_data.csv") #csv file is now converted to data frame and now we can perform operations

#check data atleast 5 columns to get insights
print(df.head())

#underrstand the data types of each column+ features+ attributes and also check for null values + missing values + number of rows and columns in the data frame
print(df.info())

#To check statistics of the data frame ex: mean, median, mode, standard deviation, min, max, etc. for numerical columns and IQR.
print(df.describe())
 #To check columns
 
print(df.columns) #we dont use() because its an attribute not a method
 #methods are functions that belong to an object and perform some action on that object. 
 #They are called using parentheses () and can take arguments. For example, df.head() is a method that returns the first few rows of the DataFrame.
 
 #Attributes are characteristics or properties of an object that hold data. 
 #They are accessed without parentheses and do not perform any action. For example, df.columns is an attribute that returns the column names of the DataFrame.
```

```python
#Data Cleaning + Feature Engineering
#Check for missing values

print(df.isnull().sum()) 
#this will give us the count of missing values in each column
#If there are missing values, we can decide how to handle them. For example, we can fill them with the mean or median of the column, or we can drop the rows with missing values.
# but we do not have any missing value in any colum and each has 0 missing value so we can proceed with the analysis without worrying about missing data.
```

```python
#For practice we will remove ome values from the data frame and then we will handle them
#Let's say we want to remove some values from the 'Sales' column
import numpy as np 
#why numpy and not pandas?
#because numpy is more efficient for handling large datasets and performing mathematical operations, while pandas is built on top of numpy and provides more functionality for data manipulation and analysis. In this case, we can use numpy to create an array of random indices to remove from the 'Sales' column.

df.loc[0, "Sales_Amount"] = np.nan 

#explanation of each part of the code:
#df: This is the DataFrame object that we are working with.
#.loc: This is a method used to access a group of rows and columns by labels or a boolean array. In this case, we are using it to access a specific cell in the DataFrame.
#[0, "Sales_amount"]: This is the label-based indexer that specifies the location of the cell we want to access. The first part (0) refers to the row index, and the second part ("Sales_amount") refers to the column name.
#= np.nan: This assigns the value of NaN (Not a Number) to the specified cell. NaN is a special floating-point value that represents missing or undefined data in pandas.
df.loc[1, "Region"] = np.nan

#Now check again if we have any null datapoints or not
print(df.isnull().sum())

#if we want to drop any exta columns we can use drop method:
#print(df.drop(columns=["Column name"], inplace=True))
```

```python
#NOW handle missing values:
#if we have numerical columns with missing values we can fill them with mean or median:
#for example if we have missing values in the 'Sales_Amount' column we can fill them with the mean of that column:

df["Sales_Amount"] = df["Sales_Amount"].fillna(df["Sales_Amount"].mean())
#for Region column we can fill the missing value with the unknown:
df["Region"] = df["Region"].fillna("Unknown")

#----Sales_Amount
#-Numeric hai
#-Mean se replace karna best

#----Region
#-Categorical hai
#-"Unknown" daal dete hain

#why we didt use dropna() method to drop the rows with missing values?
#we use dropna() method to drop the rows with missing values but in this case we have only 1 missing value in the 'Sales_Amount' column and 1 missing value in region.
#and if we drop that row we will lose that data point which can be valuable for our analysis. Moreover, if we have a large dataset with many missing values, dropping rows can lead to a significant loss of data and can also introduce bias in the analysis if the missing values are not randomly distributed.
# Therefore, it is often better to fill missing values with mean or median rather than dropping rows with missing values.

print(df.isnull().sum())
```

```python
#insights + business understanding + graphs

#Total Sales in business
total_sales=df["Sales_Amount"].sum()
print("Total sale is:", total_sales)

#Why calculate total sales?
#Overall performance measure

#total sales region wise
region_sales=df.groupby("Region")["Sales_Amount"].sum()
print("The sales by region is:", region_sales)

#Total sale by Top Sales Re--which employee earned more
top_sales_rep=df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=True)
print("The sales by top sales rep is:", top_sales_rep)
```

```python
#Profit Analysis 

#formula for profit is -- Profit = (price-cost)*units
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]
print(df[["Unit_Price", "Unit_Cost", "Quantity_Sold", "Profit"]].head())

#when there are multiple columns always use double square brackets to select them as a DataFrame instead of a Series.
#This allows us to maintain the structure of the DataFrame and perform operations on multiple columns at once.
#In this case, we are selecting the "Unit_Price", "Unit_Cost", "Quantity_Sold", and "Profit" columns to display their values for the first few rows of the DataFrame.
 
```

```python
#Visualization (Graphs)
#1. Region vs Sales (BAR CHART)
import matplotlib.pyplot as plt

region_sales = df.groupby("Region")["Sales_Amount"].sum()
region_sales.plot(kind="bar")

plt.title("Sales by Region graph")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
#comparing each regions sales
#which region is performing best
#why bar chart? to compare categories

```

```python
#Now to compare profit vs categories

import matplotlib.pyplot as plt
category_profit=df.groupby("Product_Category")["Profit"].sum()
category_profit.plot(kind="bar")

plt.title("category vs Profit graph")
plt.xlabel("category")
plt.ylabel("Profit")

plt.show()
```

```python
#Sales Trend (LINE CHART)
sales_trend = df.groupby("Sale_Date")["Sales_Amount"].sum()

sales_trend.plot(kind="line")

plt.title("Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.show()

#What it does? Same date ke saare sales ko add karta hai
# Clean trend deta hai
# Better visualization
# Business friendly
#Same data → convert ho jaata hai:

#Date        Total Sales
#1 Jan       450
#Graph becomes clean
```
